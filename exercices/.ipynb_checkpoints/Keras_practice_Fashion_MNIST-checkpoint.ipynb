```python
from tensorflow import keras 
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
```

```python
dataset = keras.datasets.fashion_mnist.load_data()
```

```python
(train_images, train_labels), (val_images, val_labels) = dataset
```

```python
train_images = train_images.astype("float32") / 255
val_images = val_images.astype("float32") / 255
```

```python
train_images = train_images.reshape((60000, 28*28)).astype("float32") / 255
val_images = val_images.reshape((10000, 28*28)).astype("float32") / 255
```

```python
sample = train_images[50:60]
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10,10))
a = 0

for i in range(3):
    for j in range(3):        
        axes[i, j].imshow(sample[a].reshape((28,28)), cmap=plt.get_cmap('gray'))
        a = a + 1

plt.show()
```

### Prepare the model

```python
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Flatten(input_shape=[28,28]),
    layers.Dense(300, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(100, activation="relu"),
    layers.Dropout(0.25),
    layers.Dense(10, activation="softmax")
])
```

```python
model.compile(optimizer="sgd",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"]
)
```

```python
epochs=30
batch_size=256
history = model.fit(
    train_images, 
    train_labels, 
    validation_split=0.2,
    batch_size=batch_size, 
    epochs=epochs
)
```

```python
val_acc = history.history["val_accuracy"]
r = range(1, epochs+1)
plt.plot(r, val_acc, "b",
         label="Validation accuracy")
plt.title("Effect of noise channels on validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
```

```python
model.evaluate(val_images, val_labels, batch_size=128)
```

```python

pd.DataFrame(history.history).plot(figsize=(15,8))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

```python
y_prob = model.predict(val_images)
y_classes = y_prob.argmax(axis=-1)
y_classes

confusion_matrix = tf.math.confusion_matrix(val_labels, y_classes)
```

```python
import seaborn as sb    
class_names = [
  "T-shirt/top",
  "Trouser",
  "Pullover",
  "Dress",
  "Coat",
  "Sandal",
  "Shirt",
  "Sneaker",
  "Bag",
  "Ankle boot"
]
# ax = plt.figure(figsize=(8, 6))
fig = sb.heatmap(confusion_matrix, annot=True, fmt='g', cmap='Greens')  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
fig.set_xlabel('Predicted labels')
fig.set_ylabel('True labels')
fig.set_title('Confusion Matrix')
fig.xaxis.set_ticklabels(class_names) 
fig.yaxis.set_ticklabels(class_names)
fig.figure.set_size_inches(10, 10)


plt.show()
```
